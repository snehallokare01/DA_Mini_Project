```python
# calculate the mean value for "Rating" column
avg_rating = df["Rating"].astype("float").mean(axis = 0)
print("Average of Rating:", avg_rating)

# replace NaN by mean value in "Rating" column
df["Rating"].replace(np.nan, avg_rating, inplace = True)
```

```python
# Clean the 'Reviews' column:
# Remove 'M' and 'k' suffixes and convert to numeric
df['Reviews'] = df['Reviews'].astype(str).str.replace('M', 'e6').str.replace('k', 'e3').astype(float)

avg_reviews = df['Reviews'].mean(axis = 0)
print("Average of Reviews:", avg_reviews)
```

```python
avg_rating = df["Rating"].astype("float").mean(axis = 0)
print("Average of Rating:", avg_rating)
```

```python
df["Reviews"].replace(np.nan, avg_reviews, inplace = True)
```

```python
df["Rating"].replace(np.nan, avg_rating, inplace = True)
```

```python
df['Installs'] = df['Installs'].astype(str).str.replace('+', '', regex=False).str.replace(',', '', regex=False).str.replace('Free', '0', regex=False)
avg_installs = df['Installs'].astype('float').mean(axis = 0)
print("Average of Installs:", avg_installs)
```

```python
avg_rating = df["Rating"].astype("float").mean(axis = 0)
print("Average of Rating:", avg_rating)
```

```python
df['Category'].value_counts()
```

```python
df['Category'].value_counts().idxmax()
```

```python
# replace the missing 'Category' values by the most frequent value
df["Category"].replace(np.nan, df["Category"].value_counts().idxmax(), inplace=True)

# simply drop whole row with NaN in "Rating" column
df.dropna(subset=["Rating"], axis=0, inplace=True)

# reset index, because we dropped rows
df.reset_index(drop=True, inplace=True)
```

```python
df.isnull().sum()
```

```python
dummy_variable_1 = pd.get_dummies(df["Category"])

df = pd.concat([df, dummy_variable_1], axis=1)

df.drop("Category", axis = 1, inplace=True)
```

```python
# Create a sample DataFrame for demonstration
sample_data = {
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'A', 'B', 'C'],
    'Value': [10, 12, 11, 15, 18, 16, 20, 22, 21, 9, 17, 23]
}
sample_df = pd.DataFrame(sample_data)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=sample_df)
plt.title('Box Plot of Values by Category (Sample Data)')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

display(sample_df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Reviews"], df["Rating"])

plt.title("Scatter Plot of Reviews vs Rating")
plt.xlabel("Reviews")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Installs"], df["Rating"])

plt.title("Scatter Plot of Installs vs Rating")
plt.xlabel("Installs")
plt.ylabel("Rating")
plt.show()
```

```python
avg_reviews_over_time = df.groupby('Last Updated')['Reviews'].mean().reset_index()

plt.figure(figsize=(14, 7))
sns.lineplot(x='Last Updated', y='Reviews', data=avg_reviews_over_time)
plt.title('Average App Reviews Over Time (by Last Updated Date)')
plt.xlabel('Last Updated Date')
plt.ylabel('Average Reviews')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.stripplot(x='Content Rating', y='Rating', data=df, jitter=True, palette='viridis')
plt.title('Strip Plot of App Ratings by Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
